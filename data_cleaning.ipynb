```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/steam_reviews 1.csv')
```

```python
df.head()
```

```python
print('The dataset contains ', df.shape[0], ' rows and ', df.shape[1], ' columns')
```

```python
print(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum()
```

```python
# Calculate the proportion of missing values to the dataset
missing_values = df['review'].isnull().sum()
size = df.shape[0]
print((missing_values/df.shape[0])*100)
```

```python
df = df.dropna(subset=['review'])
```

```python
df.info()
```

```python
df['date_posted'] = pd.to_datetime(df['date_posted'])
```

```python
df['review'] = df['review'].str.lower().str.strip()
df['title'] = df['title'].str.lower().str.strip()
```

```python
df['funny'] = pd.to_numeric(df['funny'], errors='coerce')
df['helpful'] = pd.to_numeric(df['helpful'], errors='coerce')
df['hour_played'] = pd.to_numeric(df['hour_played'], errors='coerce')
```

```python
df.groupby('recommendation')['hour_played'].mean()
```

```python
valid_labels = ['Recommended', 'Not Recommended']
df =df[df['recommendation'].isin(valid_labels)]
```
